```python
import requests
from bs4 import BeautifulSoup

needed_headers = {
    'User-Agent': "Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36"
}

response = requests.get("https://www.themoviedb.org/movie", headers=needed_headers)
page_content = response.content

soup = BeautifulSoup(page_content, "html.parser")

page_title = soup.title.text if soup.title else "Title not available"
print("\nPage Title:\t", page_title)

def create_beautiful_soup_instance(url):
    try:
        response = requests.get(url, headers=needed_headers)
        response.raise_for_status()  # Raise an exception for HTTP errors
        
        # Check for 403 status code and raise an exception
        if response.status_code == 403:
            raise requests.exceptions.HTTPError("HTTP Error 403 - Access Forbidden")

        soup = BeautifulSoup(response.content, "html.parser")
        return soup
    except requests.exceptions.HTTPError as http_err:
        print(f"\nHTTP Error:\t{http_err}")
    except requests.exceptions.RequestException as req_err:
        print(f"\nRequest Exception:\t{req_err}")
    except Exception as ex:
        print(f"\nAn error occurred:\t{ex}")
    return None

# Test cases

working_url = "https://www.themoviedb.org/movie"
forbidden_url = "https://www.example.com/forbidden"

soup_working_url = create_beautiful_soup_instance(working_url)
soup_forbidden_url = create_beautiful_soup_instance(forbidden_url)

page_title_working_url = soup_working_url.title.text if soup_working_url and soup_working_url.title else "Title not available"
print("\n\n\nWorking URL Page Title:\t", page_title_working_url)

print("\nTrying to create BeautifulSoup instance for a forbidden URL:", soup_forbidden_url)
soup_forbidden_url = create_beautiful_soup_instance(forbidden_url)
```

```python
6. Scraping the data and combining the dataframes ( 5 marks )
(a) Write a function that scrapes data (mentioned in Q5) from page number 1, 2, 3, 4 and 5 on
the URL https://www.themoviedb.org/movie and returns 5 data frames which can be exported
to csv file by calling the functions defined in Q3a, Q4c and Q5 (3 marks)
Hint : use the string “ ?page={ } “ concatenated to the string
“ https://www.themoviedb.org/movie “ which extracts the data from desired page no passed to
page value
(b) Combine the data obtained from dataframes in Q6(a) (2 marks)
Hint : refer to data integration of week4 to utilize the appropriate method available in pandas
library to combine dataframes 
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import os

def create_beautiful_soup_instance(url):
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        return BeautifulSoup(response.content, "html.parser")
    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP Error: {http_err}")
    except requests.exceptions.RequestException as req_err:
        print(f"Request Exception: {req_err}")
    except Exception as ex:
        print(f"An error occurred: {ex}")
    return None

def get_movie_titles(soup):
    movie_titles = []
    movie_elements = soup.select("h2 a[title]")
    for movie_element in movie_elements:
        movie_titles.append(movie_element["title"])
    return movie_titles

def get_movie_ratings(soup):
    movie_ratings = []
    movie_elements = soup.find_all("div", class_="user_score_chart")
    for movie_element in movie_elements:
        rating = movie_element.get("data-percent")
        if rating:
            movie_ratings.append(rating)
        else:
            movie_ratings.append("not rated")
    return movie_ratings

def get_movie_genres(movie_url):
    movie_soup = create_beautiful_soup_instance(movie_url)
    if movie_soup:
        genres_span = movie_soup.find("span", class_="genres")
        genres = [genre.text for genre in genres_span.find_all("a")]
        return genres
    return []

def get_movie_cast(movie_url):
    movie_soup = create_beautiful_soup_instance(movie_url)
    if movie_soup:
        cast_section = movie_soup.find("section", class_="panel top_billed scroller")
        cast_tags = cast_section.find_all("li", class_="card")
        cast = [cast_tag.find("p").find("a").text for cast_tag in cast_tags]
        return cast
    return []

def create_movie_dataframe(response, movie_urls):
    movie_titles = get_movie_titles(response)
    movie_ratings = get_movie_ratings(response)
    movie_genres = [get_movie_genres(movie_url) for movie_url in movie_urls]
    movie_cast = [get_movie_cast(movie_url) for movie_url in movie_urls]

    movie_data = {
        "Title": movie_titles,
        "Rating": movie_ratings,
        "Genres": movie_genres,
        "Cast": movie_cast
    }

    movie_dataframe = pd.DataFrame(movie_data)
    return movie_dataframe

def scrape_and_create_dataframes(base_url, page_range, output_folder):
    dataframes = []

    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for page_number in page_range:
        page_url = f"{base_url}?page={page_number}"
        response = create_beautiful_soup_instance(page_url)

        if response is not None:
            movie_urls = [f"https://www.themoviedb.org{movie.find('a', class_='image')['href']}" for movie in response.find_all("div", class_="card style_1")]
            movie_dataframe = create_movie_dataframe(response, movie_urls)

            csv_filename = os.path.join(output_folder, f"page_{page_number}.csv")
            movie_dataframe.to_csv(csv_filename, index=False)

            dataframes.append(movie_dataframe)

    return dataframes

base_url = "https://www.themoviedb.org/movie"
page_range = range(1, 6)
output_folder = "movie_data"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

dataframes_list = scrape_and_create_dataframes(base_url, page_range, output_folder)

for i, dataframe in enumerate(dataframes_list):
    csv_filename = os.path.join(output_folder, f"page_{page_range[i]}.csv")
    dataframe.to_csv(csv_filename, index=False)
    print(f"DataFrame {i+1} exported to {csv_filename}")

combined_dataframe = pd.concat(dataframes_list, ignore_index=True)

print(combined_dataframe)

combined_csv_filename = os.path.join(output_folder, "combined_movie_data.csv")
combined_dataframe.to_csv(combined_csv_filename, index=False)
print("Combined DataFrame exported to", combined_csv_filename)
```
